```python
import os
import shutil

def copy_images(source_folder, destination_folder, num_images):
    """Copies a specified number of images from source to destination folder."""

    # Create the destination folder if it doesn't exist
    os.makedirs(destination_folder, exist_ok=True)

    count = 0
    for filename in os.listdir(source_folder):
        if filename.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif')):  # Adjust image extensions as needed
            source_path = os.path.join(source_folder, filename)
            destination_path = os.path.join(destination_folder, filename)
            shutil.copy(source_path, destination_path)
            count += 1
            if count >= num_images:
                break

if __name__ == "__main__":
    source_folder = "/home/cm/Documents/Yash/cis583-project/stable_diff/StableDiffusion-PyTorch-main/data/CelebAMask-HQ/CelebA-HQ-img_full"
    destination_folder = "/home/cm/Documents/Yash/cis583-project/stable_diff/StableDiffusion-PyTorch-main/data/CelebAMask-HQ/firtdt"
    num_images = 20000

    copy_images(source_folder, destination_folder, num_images)
```

```python
import os
from PIL import Image

def resize_images(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith(('.jpg', '.png', '.jpeg', '.bmp')):  # Add more extensions if needed
            img_path = os.path.join(folder_path, filename)

            with Image.open(img_path) as img:
                width, height = img.size
                new_width = int(width / 2)
                new_height = int(height / 2)

                resized_img = img.resize((new_width, new_height))
                resized_img.save(img_path)  # Overwrite the original image

if __name__ == '__main__':
    folder_path = '/home/cm/Documents/Yash/cis583-project/stable_diff/StableDiffusion-PyTorch-main/data/CelebAMask-HQ/CelebA-HQ-img'  # Replace with your folder path
    resize_images(folder_path)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Read the file
with open('/home/cm/Documents/Yash/cis583-project/stable_diff/StableDiffusion-PyTorch-main/mnist/autoencoder_losses.txt', 'r') as file:
    lines = file.readlines()

# Extract the losses
recon_losses = eval(lines[0].split(': ')[1])
perceptual_losses = eval(lines[1].split(': ')[1])
# codebook_losses = eval(lines[2].split(': ')[1])

# Convert to numpy arrays for easier plotting
recon_losses = np.array(recon_losses)
perceptual_losses = np.array(perceptual_losses)
# codebook_losses = np.array(codebook_losses)

# Plot the losses
plt.figure(figsize=(10, 6))

plt.plot(recon_losses, label='Reconstruction Loss')
plt.plot(perceptual_losses, label='Perceptual Loss')
# plt.plot(codebook_losses, label='Codebook Loss')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Losses over 20 Epochs')
plt.legend()
plt.grid(True)
plt.savefig('/home/cm/Documents/Yash/cis583-project/stable_diff/StableDiffusion-PyTorch-main/celebhq/AutoEncoder_losses_mnist.png')
plt.show()
```
